```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import mplcyberpunk
from sklearn.datasets import fetch_openml
from sklearn.decomposition import PCA

plt.style.use("cyberpunk")

# Завантаження датасету
data_2d = pd.read_csv('./data/data_2d.csv', header=None)
data_mnist = pd.read_csv('./data/mnist.csv', header=None)

# Використання тільки піксельних даних для MNIST (видалення першого стовпця з мітками)
X_mnist = data_mnist.iloc[:, 1:].values

# Масштабування даних
scaler = StandardScaler()
data_2d_scaled = scaler.fit_transform(data_2d.iloc[:, 1:])
X_mnist_scaled = scaler.fit_transform(X_mnist)

def elbow_method(data):
    wcss = []  # Внутрішня сума квадратів
    for i in range(1, 11):  # Зазвичай достатньо перевірити до 10 кластерів
        kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
        kmeans.fit(data)
        wcss.append(kmeans.inertia_)

    plt.figure(figsize=(10, 5))
    plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
    plt.title('Elbow Method')
    plt.xlabel('Number of clusters')
    plt.ylabel('WCSS')
    mplcyberpunk.add_glow_effects()
    plt.show()

# Використання ліктевого методу для двох датасетів
elbow_method(data_2d_scaled)
elbow_method(X_mnist_scaled)
```

```python
# Завантаження даних
mnist = fetch_openml('mnist_784')
X = mnist.data
y = mnist.target

# Зменшення розмірності з використанням PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Кластеризація
kmeans = KMeans(n_clusters=10)
clusters = kmeans.fit_predict(X_pca)

# Візуалізація
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters, cmap='jet', s=5)
plt.colorbar()
plt.show()
```

Visualization for data_2d.csv:

```python
# Load data
X = data_2d[[1, 2]]

# Apply clustering
kmeans = KMeans(n_clusters=3)
clusters = kmeans.fit_predict(X)

# Plot
plt.scatter(X[1], X[2], c=clusters, cmap='rainbow')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('KMeans Clustering for 2D Data')
plt.show()
```

Visualization for MNIST dataset with PCA:

```python
# Assuming the first column of your data is the label and the rest are pixel values.
mnist_labels = data_mnist.iloc[:, 0]
mnist_data = data_mnist.iloc[:, 1:]

# Apply PCA
pca = PCA(n_components=2)
mnist_pca = pca.fit_transform(mnist_data)

# Visualize the result
plt.figure(figsize=(10, 8))
plt.scatter(mnist_pca[:, 0], mnist_pca[:, 1], c=mnist_labels, cmap='rainbow', s=5)
plt.colorbar()
plt.title('MNIST PCA Visualization')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

Conclusion

When you visualize datasets using techniques like PCA, the visualizations can offer insights into the structure and relationships in the data. I will provide a general conclusion for the 2D dataset (which we haven't seen) and the MNIST dataset based on the PCA visualization:

For the 2D dataset:

1. Assuming the 2D dataset refers to some dataset with two dimensions/features, visualization in a 2D space provides a direct representation of the data distribution.
2. Clusters or groupings of data points may indicate regions of similar data points or categories. If these are labeled, the label distribution within these clusters can be observed directly.
3. Outliers can be easily spotted if data points are far removed from any clusters or groupings.

For the MNIST dataset:

1. When the MNIST dataset is reduced to 2D using PCA and visualized, the main goal is to see how well-separated the different digits (0 through 9) are in this reduced space.
2. If clusters or groupings are visible, it means that digits have distinct patterns that PCA could capture even in just two dimensions.
3. Some overlap between clusters is expected because PCA is linear, and some information is inevitably lost when reducing from 784 dimensions (28x28 pixels) to 2. Overlaps might indicate that certain digits have similar patterns or structures.
4. The variation captured by the two principal components provides an idea of how much "information" from the original dataset is retained in the 2D visualization. The more variation captured, the more accurate our 2D representation is compared to the original 784-dimensional data.

In conclusion, PCA visualizations, especially for high-dimensional datasets like MNIST, offer a simplified view to gain insights into the data. It's essential to remember that while these visualizations are informative, they don't capture the entire complexity of the original data.
